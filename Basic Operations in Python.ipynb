## Python Functions and Basic operations (without using any libraries)

```python
def is_word_there(text, word):
    
    """ 
    return return 1 if word is in the text.
    return -1 if word is not in the text
    """
    #read file and store in variable 'data'
    with open(text, 'r') as file:
        data = file.read().lower()
    
    #check if 'word' exists in the data
    if word.lower() in data:
        return 1
    return -1
    
def count_word(text, word):
    
    """
    return the number of word counts in a text.
    if the word does not in the text, return -1.
    """
    #read file and store in varibale data
    data = open(text).read().lower()
    #count the number of times 'word' appears in 'data'
    count = data.count(word.lower())
    if(count > 0):
        return count
    return -1

def get_sum(numbers):
    
    """
    given a list of numbers. find the sum
    """
    count = 0
    #iterating over all values in numbers
    for i in numbers:
        count += i

    return count
    
def get_factorial(number):
    
    """
    given a number, find its factorial
    """
    factorial = 1
    while(number > 1):
        factorial = factorial * number
        number -= 1

    return factorial
    
def sum_lists(list1, list2):
    
    """
    given that list1 and list 2 are equal sizes.
    get the sum of these two lists elementwise
    for example:
    list1 = [1, 6, 12]
    list2 = [4, 2, 14]
    sum
    [5, 8, 26]
    """
    result = []
    
    for i in range(len(list1)):
        result.append(list1[i] + list2[i])

    return result
    
def isEven(number):  

    """
    determine if a number is even
    if even, return +1
    if odd, return -1 
    """
    #number is even if after dividing it by 2, remainder is 0
    if(number % 2 == 0):
        return 1
    else:
        return -1

def isOdd(number):
        
    """
    determine if a number is odd
    if odd, return +1
    if not return -1
    
    """
     #isOdd function works in a opposite fashion to isEven. Here instead of re-writing the code, using previously defined function
    if isEven(number) == -1:
        return 1
    else:
        return -1
    
```
